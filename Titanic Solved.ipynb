```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn import tree
from sklearn.metrics import accuracy_score

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
dataset = pd.read_csv('train.csv')
X = dataset.iloc[:,2:13]
Y = dataset['Survived']
#X.info()
#X.head(10)
#dataset.head(10)
#dataset.columns
dataset
```

```python
#missing datas
# Age, Cabin, Embark

"""
#numeric
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy = 'mean')
imputer.fit(X.iloc[:, 3:4])
X.iloc[:, 3:4] = imputer.transform(X.iloc[:, 3:4])
#Object
"""

X['Age'] = X['Age'].fillna(-0.5)
X['Embarked'] = X['Embarked'].fillna('S')
X = X.drop('Cabin', axis=1)
X = X.drop('Ticket', axis = 1)
X = X.drop('Name', axis=1)
X.head(10)
```

```python
#Encode

ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,6])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

#Encode Dependent Variable
le = LabelEncoder()
le.fit_transform(Y)
```

```python
#Split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state= 42)
```

```python
# Standardization

sc = StandardScaler()
X_train[:, [0,2,3,4,5]] = sc.fit_transform(X_train[:, [0,2,3,4,5]])
X_test[:, [0,2,3,4,5]] = sc.fit_transform(X_test[:, [0,2,3,4,5]])
```

```python
#model

clf = tree.DecisionTreeClassifier()
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
acc=accuracy_score(Y_test,Y_pred)
print(f"The accuracy of this model is {acc * 100} perc%ntage")
```

```python
#model

clf= RandomForestClassifier(n_estimators=100)
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
accur = metrics.accuracy_score(Y_test,Y_pred)
print(f"The accuracy of this model is {accur * 100} perc%ntage")
```
